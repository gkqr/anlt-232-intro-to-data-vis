## Introduction
Effective data visualization isn't just about displaying data but also about making it **clear, aesthetically pleasing, and easy to interpret**. Seaborn builds on Matplotlib‚Äôs capabilities by offering **built-in themes and styles** that improve readability with minimal effort.

In this lesson, we'll explore how Seaborn‚Äôs styling functions allow us to adjust **color themes, background styles, and contextual settings** to create well-designed plots.

## Loading the Dataset
To demonstrate Seaborn's figure styles and themes, we'll use the `tips` dataset, which contains information on restaurant bills, tips, and customer demographics.

```python
import seaborn as sns
import matplotlib.pyplot as plt  # Import Matplotlib for further customization

# Sample data
tips = sns.load_dataset("tips")
tips.head()
```

## Setting Styles in Seaborn
Seaborn provides **five built-in** themes to customize the appearance of plots:

- `darkgrid` (default) ‚Äì A grid with a dark background

- `whitegrid` ‚Äì A grid with a light background

- `dark` ‚Äì Dark background without gridlines

- `white` ‚Äì White background without gridlines

- `ticks` ‚Äì White background with ticks emphasized

You can change the style using `sns.set_style()`:

```python
# Apply a style
# See: https://seaborn.pydata.org/generated/seaborn.set_style.html
sns.set_style("whitegrid")

# Create a scatter plot of total_bill vs tip with custom style
sns.scatterplot(
    data=tips,  # Specifies the tips dataset as the data source
    x="total_bill",  # Uses the total bill column as the x-axis
    y="tip",  # Uses the tip column as the y-axis
)

# Adds a descriptive title
plt.title("Scatter Plot with 'whitegrid' Style")

# Shows the plot
plt.show()
```

## Customizing Figure Aesthetics
Seaborn allows further control over individual plot elements using `sns.set_style()` with `rc` parameters. This provides fine-grained control over **grid visibility, axis colors, and more**.

**Example:** Removing the top and right spines (borders of the plot):

```python
# Apply a style including background and spine customization
sns.set_style("whitegrid", {"axes.spines.right": False, "axes.spines.top": False})

sns.scatterplot(
    data=tips,  # Specifies the tips dataset as the data source
    x="total_bill",  # Uses the total bill column as the x-axis
    y="tip",  # Uses the tip column as the y-axis
)

# Adds a descriptive title
plt.title("Scatter Plot with Customized Spines")

# Shows the plot
plt.show()
```

## Changing Plot Context
Seaborn includes a `set_context()` function that adjusts font size and element scaling based on the **intended use of the plot**. The available contexts are:

- `paper` ‚Äì For small plots in reports or articles

- `notebook (default)` ‚Äì For standard Jupyter Notebook visuals

- `talk` ‚Äì Enlarged text for presentations

- `poster` ‚Äì Even larger text for posters

**Example:**

```python
# TODO: Try replacing "talk" with other contexts to see how they affect the plot.
# See: https://seaborn.pydata.org/generated/seaborn.set_context.html
sns.set_context("talk")

sns.scatterplot(
    data=tips,  # Specifies the tips dataset as the data source
    x="total_bill",  # Uses the total bill column as the x-axis
    y="tip",  # Uses the tip column as the y-axis
)

# Adds a descriptive title
plt.title("Scatter Plot with 'talk' Context")

# Shows the plot
plt.show()
```

You can also fine-tune individual parameters:

```python
# Adjust the font scale
sns.set_context("talk", font_scale=1.5)

sns.scatterplot(
    data=tips,  # Specifies the tips dataset as the data source
    x="total_bill",  # Uses the total bill column as the x-axis
    y="tip",  # Uses the tip column as the y-axis
)

# Adds a descriptive title
plt.title("Scatter Plot with 'talk' Context")

# Shows the plot
plt.show()
```

## Using Seaborn‚Äôs Color Palettes
Seaborn provides built-in color palettes that can be applied easily. To see available palettes:

```python
# See: https://seaborn.pydata.org/generated/seaborn.color_palette.html
sns.color_palette()
```

To apply a color palette:

```python
sns.set_context("notebook")

# Set the color palette
# See: https://seaborn.pydata.org/generated/seaborn.set_palette.html
sns.set_palette("pastel")

sns.scatterplot(
    data=tips,  # Specifies the tips dataset as the data source
    x="total_bill",  # Uses the total bill column as the x-axis
    y="tip",  # Uses the tip column as the y-axis
)

# Adds a descriptive title
plt.title("Scatter Plot with 'pastel' Palette")

# Shows the plot
plt.show()
```

Other common palettes:

- `"deep"`

- `"muted"`

- `"bright"`

- `"dark"`

- `"colorblind"`
  
You can also define a custom color palette:

```python
# Set custom palette
custom_palette = ["#3498db", "#e74c3c", "#2ecc71"]
sns.set_palette(custom_palette)

sns.scatterplot(
    data=tips,  # Specifies the tips dataset as the data source
    x="total_bill",  # Uses the total bill column as the x-axis
    y="tip",  # Uses the tip column as the y-axis
)

# Adds a descriptive title
plt.title("Scatter Plot with Custom Palette")

# Shows the plot
plt.show()
```
