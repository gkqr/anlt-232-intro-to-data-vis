## Introduction
Understanding relationships between variables is a fundamental part of data analysis. Seaborn provides **relational plots** to help visualize how two numerical variables interact, with options to incorporate categorical variables for deeper insights.

In this lesson, we'll explore Seaborn‚Äôs **relational plotting functions** and their applications in uncovering trends and patterns in data.

## Types of Relational Plots in Seaborn
Seaborn offers the `relplot()` function, a high-level interface for relational plots, with two key types:

| Plot Type | Description | 
|----------|----------|
| **Scatter Plot** (`sns.scatterplot`) | Displays individual data points and relationships between two numerical variables.   | 
| **Line Plot** (`sns.lineplot`)  | 	Shows trends over a continuous variable, such as time series data.  | 

Both plots allow customization with **hue**, **size**, and **style** parameters to add categorical distinctions.

## Loading the Dataset
For this lesson, we'll use the `tips` dataset, which contains information on restaurant bills, tips, and customer demographics.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Tips dataset
tips = sns.load_dataset("tips")

# Display the first few rows
tips.head()
```

## Scatter Plot
A **scatter plot** is useful for identifying patterns, correlations, or clusters between two numerical variables. It plots individual data points, where the x-axis represents one variable, and the y-axis represents another.

In the following example, we visualize the relationship between **total bill amount** and **tip amount**:

```python
# Sets the current figure size
plt.figure(figsize=(8, 5))

# See: https://seaborn.pydata.org/generated/seaborn.scatterplot.html
sns.scatterplot(
    data=tips,  # Specifies the tips dataset as the data source
    x="total_bill",  # Uses the total bill column as the x-axis
    y="tip",  # Uses the tip column as the x-axis 
    hue="day",  # Points are color coded based on the day of the week
    style="sex",  # Different marker styles (e.g., circles and squares) distinguish between male and female customers
    palette="muted",  # Provides a subtle, aesthetically pleasing color scheme
)

# Adds a descriptive title
plt.title("Total Bill vs. Tip Amount")

# Shows the plot
plt.show()
```

This plot helps us analyze whether **higher bills lead to higher tips** and if **tipping behavior varies by day or gender**.

### Adding More Information
We can use additional arguments to encode more dimensions:

```python
# Sets the current figure size
plt.figure(figsize=(8, 5))

sns.scatterplot(
    data=tips,  # Specifies the tips dataset as the data source 
    x="total_bill",  # Uses the total bill column as the x-axis
    y="tip",  # Uses the tip column as the x-axis 
    hue="smoker",  # Colors points based on whether the customer is a smoker or non-smoker
    style="time",  # Uses different markers for lunch vs. dinner customers
    size="size",  # Adjusts the marker size based on the party size
    palette="coolwarm",  # Sets the color palette
)

# Adds a descriptive title
plt.title("Total Bill vs. Tip Amount (with Additional Encodings)")

# Adds x- and y-labels
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip Amount ($)")

# Shows the plot
plt.show()
```

This version of the scatter plot introduces **three additional categorical variables**, making it even more informative.

## Line Plot
A **line plot** is useful for analyzing trends in continuous data, particularly over time. Instead of plotting individual points, it connects them to show a pattern.

In the example below, we examine how the **total bill amount** varies by **day**:

```python
# Sets the current figure size
plt.figure(figsize=(15, 6))

# See: https://seaborn.pydata.org/generated/seaborn.lineplot.html
sns.lineplot(
    data=tips,  # Specifies the tips dataset as the data source
    x="day",  # Uses the day column as the x-axis (a unit of time)
    y="total_bill",  # Uses the total bill column as the numerical y-axis
    hue="size",  # Lines are colored based on the size of the party
    style="sex",  # Different marker styles (e.g., circles and squares) distinguish between male and female customers
    markers=True,  # Displays points at data locations along the lines
    dashes=False,  # Uses solid lines instead of dashed lines
    errorbar=None,  # Suppresses the measure of variability (default = 95% confidence interval (CI))
    palette="tab10",  # Sets the color palette
)

# Adjusts legend location and style
plt.legend(loc="lower right", ncol=3)

# Adds a descriptive title
plt.title("Total Bill Trends by Day")

# Adds x- and y-labels
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")

# Shows the plot
plt.show()
```

This plot helps us identify whether **larger groups tend to spend more money** and if **spending patterns vary by day or gender**.

## Using `relplot()` for Faceted Relational Plots
Seaborn‚Äôs `relplot()` function allows us to create faceted plots, where multiple subplots are generated based on a categorical variable.

```python
# See: https://seaborn.pydata.org/generated/seaborn.relplot.html
sns.relplot(
    data=tips,  # Specifies the tips dataset as the data source
    x="total_bill",  # Uses the total bill column as the x-axis
    y="tip",  # Uses the tip column as the x-axis 
    hue="sex",  # Points are colored by gender
    col="day",  # A separate subplot is created for each day of the week
    kind="scatter",  # Specifies that we want scatter plots
    height=4,  # Adjusts subplot size
    aspect=1,  # Adjusts subplot size
)

# Shows the plot
plt.show()
```

This approach is **useful for comparing trends across different groups**, as it allows us to see how relationships vary across days without cluttering a single plot.
