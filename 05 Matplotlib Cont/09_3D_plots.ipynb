# 3D Plots

Another way to plot the function from the previous section is to plot a surface on a 3D perspective plot.  
Recall the function we just looked at:
$$
z = f(x,y) = \frac{\sin(x \cdot y)}{1 + x^2 + y^2}
$$

We represented this function with a **contour** plot like this: 

```python
# these are done in my startup file
# import matplotlib.pyplot as plt
# import numpy as np

def f(x, y):
    return np.sin(x*y)/(1+ x**2 + y**2)
```

```python
# Create arrays of 50 equally spaced values from -2 to 2 for the x and y
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)

# Create a meshgrid from the x and y arrays, which generates coordinate matrices for vectorized evaluation
X, Y = np.meshgrid(x, y)

# Evaluate the function f at every point in the meshgrid to produce the Z values
Z = f(X, Y)
```

```python
# Plot filled contour lines
# See: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contourf.html
plt.contourf(X, Y, Z, 20, cmap='RdGy')

# Add a color bar
plt.colorbar()

# Display the figure
plt.show()
```

## Wireframes and Surface Plots

Two other types of 3D plots which work on *gridded data* are **wireframe** and **surface** plots.  
These take a grid of values and project it onto the specified three-dimensional surface, and can make the resulting three-dimensional forms quite easy to visualize.

### Three-dimensional Plotting in Matplotlib

*Matplotlib* was designed to be a *two-dimensional* plotting library.  
Around the time of the 1.0 release, some 3D plotting utilities were built on top of matplotlib’s 2D display,
and the result is a convenient (if somewhat limited) set of tools for *three-dimensional* data visualization.

A three-dimensional axes can be created by passing the keyword `projection='3d'` to any of the normal axes creation routines:

```python
# When I do 3D plots, I always turn on %matplotlib widget so I can interact with figure via the mouse
%matplotlib widget

# Create a Figure
fig = plt.figure()

# Tell the Axes you want a "3D projection" perspective
ax = plt.axes(projection='3d')

# Plot a 3D wireframe using the X, Y, and Z data
# See: https://matplotlib.org/stable/plot_types/3D/wire3d_simple.html
ax.plot_wireframe(X, Y, Z)

# Show the figure
plt.show()
```

A **surface plot** is like a wireframe plot, but each face of the wireframe is a filled polygon.

Adding a colormap to the filled polygons can aid perception of the topology of the surface being visualized:

```python
%matplotlib widget

# Create a Figure
fig = plt.figure()

# Tell the Axes you want a "3D projection" perspective
ax = plt.axes(projection='3d')

# Plot a 3D surface plot using the X, Y, and Z data
# See: https://matplotlib.org/stable/plot_types/3D/surface3d_simple.html
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='RdGy', edgecolor='black')

# Show the figure
plt.show()
```

Or, with `edgecolor='none'`.

```python
%matplotlib widget

# Create a Figure
fig = plt.figure()

# Tell the Axes you want a "3D projection" perspective
ax = plt.axes(projection='3d')

# Plot a 3D surface plot using the X, Y, and Z data with no edgecolor
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='RdGy', edgecolor='none')

# Show the figure
plt.show()
```

Compare this with the **contour** plot above.

## `plt.contour3D()`

Analogous to the contour plots we talked about above, Matplotlib contains tools to create three-dimensional relief plots using the same inputs.  

Like two-dimensional `ax.contour()` plots, `ax.contour3D()` requires all the input data to be in the
form of two-dimensional regular grids, with the **Z** data evaluated at each point.

```python
%matplotlib widget

# Create a Figure
fig = plt.figure()

# Tell the Axes you want a "3D projection" perspective
ax = plt.axes(projection='3d')

# Create a 3D contour plot with 20 contour levels
ax.contour3D(X, Y, Z, 20, cmap='RdGy')

# Just for practice:
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
# Notice, we are here using the API forms: set_xlabel(), etc.

plt.show()
```

## 3D Line and Scatter Plots

The most basic 3D plot is a line or collection of scatter plot created from sets of $(x, y, z)$ triples.  
In analogy with the more common two-dimensional discussed earlier, these
can be created using the `ax.plot3D()` and `ax.scatter3D()` functions.  
The call signature for these is nearly identical to that of their two-dimensional counterparts.

Here is how you make a **3D Line Plot**

```python
%matplotlib widget

# Create a Figure
fig = plt.figure()

# Tell the Axes you want a "3D projection" perspective
ax = plt.axes(projection='3d')

# Data for a 3D line: a set of X, Y, and Z arrays. ax.plot3D() will connect these "dots" with lines.
zline = np.linspace(0, 15, 20)
xline = np.sin(zline)
yline = np.cos(zline)

# Plot it
ax.plot3D(xline, yline, zline, 'red')

# Show the figure
plt.show()
```

Similarly, we can make a **3D Scatter Plot**

```python
%matplotlib widget

# Set the random seed to ensure reproducibility of random numbers   
np.random.seed(666)

# Data for 3D scattered points, same pattern as above, with some randomness
zdata = 15 * np.random.random(200)
xdata = np.sin(zdata) + 0.5 * np.random.randn(200)
ydata = np.cos(zdata) + 0.5 * np.random.randn(200)

# Create a Figure
fig = plt.figure()

# Tell the Axes you want a "3D projection" perspective
ax = plt.axes(projection='3d')

# Plot it
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens')

# Show the figure
plt.show()
```

Above, we used the **c=** (color) keyword to color the points according the the `zdata` (height), using the `'Greens'` colormap.

-------------


You can also combine **3D Line** and **3D Scatter** plots, for an effective construction.

```python
%matplotlib widget

# Create a Figure
fig = plt.figure()

# Tell the Axes you want a "3D projection" perspective
ax = plt.axes(projection='3d')

# Data for a 3D line: a set of X, Y, and Z arrays. ax.plot3D() will connect these "dots" with lines.
zline = np.linspace(0, 15, 20)
xline = np.sin(zline)
yline = np.cos(zline)

# Plot it
ax.plot3D(xline, yline, zline, 'red')
ax.scatter3D(xline,yline,zline, color='black')

# Show the figure
plt.show()
```
