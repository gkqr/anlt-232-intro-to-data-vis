## What is Tableau?
[**Tableau**](https://www.tableau.com/) is a leading data visualization and business intelligence tool that enables users to analyze data and generate insights through interactive dashboards and reports. Unlike coding-based tools such as Dash, Tableau provides a **drag-and-drop interface** that allows users to build visualizations with minimal programming knowledge.

![Intro to Tableau](images/intro_to_tableau.png)

**Tableau** is widely used in industries such as finance, healthcare, marketing, and operations due to its ability to quickly process large datasets and create compelling data stories. Organizations use Tableau for tasks such as:

- **Sales and revenue analysis** ‚Äì Tracking company performance and trends.

- **Customer segmentation** ‚Äì Identifying key customer groups based on behaviors.

- **Supply chain optimization** ‚Äì Monitoring inventory and logistics in real time

### Key Benefits of Tableau
1. **User-friendly Interface**: Drag-and-drop functionality makes it accessible to non-programmers.

2. **High-Speed Data Processing** ‚Äì Can handle large datasets efficiently.

3. **Data Connectivity** ‚Äì Supports connection to various data sources such as Excel, SQL databases, Google Sheets, and cloud-based data warehouses.

4. **Interactive Dashboards** ‚Äì Allows for filtering, drill-downs, and dynamic visualizations.

5. **Enterprise Scalability** ‚Äì Can be deployed at an organizational level using Tableau Server or Tableau Cloud.

### Tableau Product Offerings
There are multiple versions of Tableau to cater to various user needs:

| Tableau Product | Description | Best For |
|----------|----------|----------|
| **Tableau Public** | 	Free version for creating and sharing visualizations online | Beginners and students | 
| **Tableau Desktop** | Paid version with full functionality and the ability to save work locally | Analysts and professionals | 
| **Tableau Server** | Enterprise solution for hosting and sharing dashboards within an organization | Large businesses | 
| **Tableau Cloud** | Cloud-hosted version of Tableau Server for remote access and collaboration | Teams working across different locations | 

üí° *Note: For this course, we'll use Tableau Public, a free version suitable for learning and sharing visualizations.*

## What is Tableau Public?
**Tableau Public** is a free platform for creating, publishing, and sharing interactive visualizations with a global audience. It‚Äôs an excellent way to learn Tableau, showcase your work, and explore dashboards created by others‚Äîall without needing a paid license. 

However, keep in mind that everything shared on Tableau Public is accessible to everyone. Make sure any data you publish is intended for <u>public consumption</u>! If you need to work with private data, consider [Tableau‚Äôs other offerings](https://www.tableau.com/products).

## Exploring the Tableau Interface
Once installed, launch **Tableau Public** to familiarize yourself with its interface.

### Key Components of the Tableau Interface

| Component | Description |
|----------|----------|
| **Start Page** | Provides access to recent workbooks and data connection options. | 
| **Data Pane** | Displays connected data fields, categorized as **Dimensions** (categorical) and **Measures** (numerical). |

| **Show Me Panel** | Recommends visualization types based on selected data fields. | 
| **Dashboard & Story Tabs** | Combine multiple visualizations into interactive dashboards or guided narratives. |

## Understanding Dimensions vs. Measures
Tableau automatically classifies data fields into two categories:

- <span style="color: blue;">**Dimensions (Blue fields)**</span> ‚Üí Categorical data that defines groups (e.g., `Customer Name`, `City`, `Product Category`).

- <span style="color: green;">**Measures (Green fields)**</span> ‚Üí Numerical data that can be aggregated (e.g., `Sales`, `Profit,`, `Quantity`).

üí° *You can manually adjust a field‚Äôs classification by dragging it between **Dimensions** and **Measures**.*

## Saving and Publishing Your Work
Tableau Public requires users to publish their work online, as local saving is not supported.

### Steps to Save and Publish:
1. Click **File > Save to Tableau Public As‚Ä¶**
   
2. Log in to your **Tableau Public account**.

3. Enter a **title** for your visualization and click **Save**.

4. Your dashboard will be uploaded and accessible online.

üí° *If using **Tableau Desktop**, you can save your work locally as a `.twbx` file.*
